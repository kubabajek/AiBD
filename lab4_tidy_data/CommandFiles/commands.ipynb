```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 300)
```

**Wczytanie oryginalnych danych** <br>
Usunąłem wszystkie znaki niebędące danymi, które miały za zadanie je rozdzielić (OI, OS, SI, I, S, B, D)


```python
df = pd.read_csv('../OriginalData/weather.txt', header=None)

#Dane z pominięciem pierwszej kolumny, która wymaga rozdzielenia
data = df[0].map(lambda x: x[21:]) 
symbols = ["OI","OS","SI","I","S","B","D"]

for s in symbols:
    data = data.str.replace(f"{s}", " ")
    
data=data.map(str.lstrip)
data=data.str.split(r"\s+", expand=True)
data=data[list(range(31))]
data=data.rename(columns={i: f"d{i+1}" for i in range(31)})
data=data.replace("-9999", np.NaN)
for i in range(1,32):
    data[f"d{i}"] = data[f"d{i}"].astype(float)/10

#Stworzenie nowej tabeli na podstawie danych
weather = pd.DataFrame()
weather['station'] = df[0].map(lambda x: x[:11])
weather['year'] = df[0].map(lambda x: x[11:15]).astype(int)
weather['month'] = df[0].map(lambda x: x[15:17]).astype(int)
weather['type'] = df[0].map(lambda x: x[17:21]).str.lower()
for i in range(1,32):
    weather[f"d{i}"] = data[f"d{i}"]
weather
```

**Messy Data** <br>
Wybranie tylko pomiarów z 2010r.

```python
weather2010 = weather[(weather["year"] == 2010)].reset_index(drop=True)
weather2010
```

**Molten Data** <br>
Zebrałem dni z kolumn do wierszy, utworzyłem datę i usunąłem puste wartości

```python
melt = pd.melt(weather2010, id_vars=["station", "year", "month", "type"], var_name="day")
melt["day"] = melt["day"].map(lambda x: x[1:]).astype(int)
melt["date"] = pd.to_datetime(melt[["year", "month", "day"]], errors="coerce")
melt = melt[["station", "date", "type", "value"]]
melt = melt.dropna()
melt = melt.sort_values(["station", "date", "type"])
melt=melt.reset_index(drop=True)
melt
```

**Tidy Data** <br>
Wyciągnąłem typ pomiaru jako kolumny i rozdzieliłem na 2 tabele, jedna z temperaturami (tmax, tmin), druga z opadami (prcp)

```python
tidy = melt.set_index(["station", "date", "type"]).unstack()
tidy.columns = tidy.columns.droplevel(0)
tidy.columns.name = None
tidy = tidy.reset_index()

temperature = tidy[["station","date","tmin","tmax"]]
temperature = temperature.dropna()
temperature.to_csv('../AnalysisData/temperature.csv', index=False)

precipitation = tidy[["station","date","prcp"]]
precipitation = precipitation.dropna()
precipitation.to_csv('../AnalysisData/precipitation.csv', index=False)
```
