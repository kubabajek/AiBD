W zadaniu zaimportowałem potrzebne pakiety (numpy i matplotlib). Następnie zdefiniowałem zadaną funkcję. W drugiej komórce stworzyłem trzy zakresy argumentów i zapisałem wyniki funkcji dla nich do zmiennych. Następnie wyświetliłem wykresy wszystkich trzech wyników dodając odpowiednie opisy.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def fun(x):
    return x**2+5
```

```python
x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(0,5)

result1 = fun(x1)
result2 = fun(x2)
result3 = fun(x3)

plt.plot(x1,result1,label='Pierwszy zakres')
plt.title('Pierwszy wykres')
plt.xlabel('Liczba x')
plt.ylabel('Wynik funkcji')
plt.legend()
plt.show()

plt.plot(x2,result2,label='Drugi zakres')
plt.title('Drugi wykres')
plt.xlabel('Liczba x')
plt.ylabel('Wynik funkcji')
plt.legend()
plt.show()

plt.plot(x3,result3,label='Trzeci zakres')
plt.title('Trzeci wykres')
plt.xlabel('Liczba x')
plt.ylabel('Wynik funkcji')
plt.legend()
plt.show()
```
